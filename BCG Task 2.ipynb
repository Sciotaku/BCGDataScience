## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

```python
# Distribution of churn
plt.figure(figsize=(8, 6))
sns.countplot(data=client_df, x='churn')
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
# Stacked bar plot of churn by channel sales
plt.figure(figsize=(12, 8))
plot_stacked_bars(client_df.groupby(['channel_sales', 'churn']).size().unstack(), 'Churn by Channel Sales')
```

```python
# Stacked bar plot of churn by has_gas
plt.figure(figsize=(8, 6))
plot_stacked_bars(client_df.groupby(['has_gas', 'churn']).size().unstack(), 'Churn by Has Gas')
```

```python
# Distribution of cons_12m for churned and retained companies
fig, ax = plt.subplots(figsize=(10, 6))
plot_distribution(client_df, 'cons_12m', ax)
plt.title('Distribution of cons_12m for Churned and Retained Companies')
plt.show()
```

```python
# Distribution of net_margin for churned and retained companies
fig, ax = plt.subplots(figsize=(10, 6))
plot_distribution(client_df, 'net_margin', ax)
plt.title('Distribution of net_margin for Churned and Retained Companies')
plt.show()
```

```python
# Calculate price sensitivity
client_df['price_sensitivity'] = client_df['forecast_price_energy_off_peak'] + client_df['forecast_price_energy_peak'] + client_df['forecast_price_pow_off_peak'] - (price_df['price_off_peak_var'] + price_df['price_peak_var'] + price_df['price_mid_peak_var'])

# Check the distribution of price sensitivity
plt.figure(figsize=(10, 6))
sns.histplot(data=client_df, x='price_sensitivity', kde=True)
plt.title('Distribution of Price Sensitivity')
plt.xlabel('Price Sensitivity')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of price sensitivity and churn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=client_df, x='price_sensitivity', y='churn')
plt.title('Price Sensitivity vs. Churn')
plt.xlabel('Price Sensitivity')
plt.ylabel('Churn')
plt.show()

# Calculate correlation coefficient between price sensitivity and churn
correlation = client_df['price_sensitivity'].corr(client_df['churn'])
print("Correlation coefficient between price sensitivity and churn:", correlation)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=client_df, x='churn', y='price_sensitivity')
plt.title('Price Sensitivity by Churn')
plt.xlabel('Churn')
plt.ylabel('Price Sensitivity')
plt.show()
```
